```python
import time
import random

def Odd_even_sort(l, n) : # odd even sort function
    is_sorted = 0 # 정렬 완료 플래그
    while is_sorted == 0 : # 전체 정렬 while문
        is_sorted = 1
        temp = 0
        for i in range(0, n - 1, 2) : # odd phase (2씩 증가)
            if l[i] > l[i + 1] : # Bubble sort
                temp = l[i + 1]
                l[i + 1] = l[i]
                l[i] = temp                
                is_sorted = 0
                
        for i in range(1, n - 1, 2) : # even phase (2씩 증가)
            if l[i] > l[i + 1] : # Bubble sort
                temp = l[i + 1]
                l[i + 1] = l[i]
                l[i] = temp
                is_sorted = 0
    return        

def Checksort(a, n): # 정렬 체크 함수
    sorted = True
    
    for i in range(1, n):
        if a[i - 1] > a[i] :
            sorted = False
        if not sorted :
            break
    if(sorted):
        print("정렬 완료")
    else :
        print("정렬 오류!")


In_l1 = []
In_l2 = []
In_l3 = []

for i in range(0, 1000): # 난수로 리스트 생성
  number =random.randint(0, 1000)
  In_l1.append(number)
    
for i in range(0, 5000): # 난수로 리스트 생성
  number =random.randint(0, 1000)
  In_l2.append(number)
    
for i in range(0, 10000): # 난수로 리스트 생성
  number =random.randint(0, 1000)
  In_l3.append(number)
    
start_time1 = time.time()

Odd_even_sort(In_l1, 1000)

end_time1 = time.time()

start_time2 = time.time()

Odd_even_sort(In_l2, 5000)

end_time2 = time.time()

start_time3 = time.time()

Odd_even_sort(In_l3, 10000)

end_time3 = time.time()

print("입력 :", 1000)
print("odd even sort time : ", format(end_time1 - start_time1, 'f'), "sec") # 시간 출력(단위 : 초)
Checksort(In_l1, 1000)

print("입력 :", 5000)
print("odd even sort time : ", format(end_time2 - start_time2, 'f'), "sec") # 시간 출력(단위 : 초)
Checksort(In_l2, 5000)

print("입력 :", 10000)
print("odd even sort time : ", format(end_time3 - start_time3, 'f'), "sec") # 시간 출력(단위 : 초)
Checksort(In_l3, 10000)
```
