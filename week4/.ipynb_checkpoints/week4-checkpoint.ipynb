```python
import time
import random

def Shellsort_1(l) : # 쉘 정렬 함수
    n = len(l) # 총 개수
    h = 1
    while h < n : # 첫번째 h 계산
        h = 3 * h + 1 # 증가식
    while h > 0 : # Gap으로 만들어진 부분 배열에서 삽입정렬 실행
        for i in range(h, n) :
            temp = l[i]
            j = i - h
            while j >= 0 and l[j] > temp : # 삽입정렬
                l[j + h] = l[j]
                j -= h # 1gap 씩 줄여가며 비교
            l[j + h] = temp
        h //= 3 # 감소식
        
def Shellsort_2(l) : # 쉘 정렬 함수
    n = len(l) # 총 개수
    h = 1
    while h < n : # 첫번째 h 계산
        h = 2 * h + 1 # 증가식
    while h > 0 : # Gap으로 만들어진 부분 배열에서 삽입정렬 실행
        for i in range(h, n) :
            temp = l[i]
            j = i - h
            while j >= 0 and l[j] > temp : # 삽입정렬
                l[j + h] = l[j]
                j -= h # 1gap 씩 줄여가며 비교
            l[j + h] = temp
        h //= 2 # 감소식
        
def Shellsort_3(l) : # 쉘 정렬 함수
    n = len(l) # 총 개수
    h = 1
    while h < n : # 첫번째 h 계산
        h = 5 * h + 1 # 증가식
    while h > 0 : # Gap으로 만들어진 부분 배열에서 삽입정렬 실행
        for i in range(h, n) :
            temp = l[i]
            j = i - h
            while j >= 0 and l[j] > temp : # 삽입정렬
                l[j + h] = l[j]
                j -= h # 1gap 씩 줄여가며 비교
            l[j + h] = temp
        h //= 5 # 감소식
        
def Checksort(a, n): # 정렬 체크 함수
    sorted = True
    for i in range(1, n):
        if a[i - 1] > a[i] :
            sorted = False
        if not sorted :
            break
    if(sorted):
        print("정렬 완료")
    else :
        print("정렬 오류!")

n = 10000000 # 입력

In_l1 = []
In_l2 = []
In_l3 = []

for i in range(0, n): # 난수로 리스트 생성
  number =random.randint(0, n)
  In_l1.append(number)

for i in range(0, n): # 난수로 리스트 생성
  number =random.randint(0, n)
  In_l2.append(number)

for i in range(0, n): # 난수로 리스트 생성
  number =random.randint(0, n)
  In_l3.append(number)


start_time1 = time.time()

Shellsort_1(In_l1)

end_time1 = time.time()

start_time2 = time.time()

Shellsort_2(In_l2)

end_time2 = time.time()

start_time3 = time.time()

Shellsort_3(In_l3)

end_time3 = time.time()

Checksort(In_l1, n)
print("증가식 : 3h + 1, 감소식 : h/3, N :", n)
print("Shell sort time : ", format(end_time1 - start_time1, 'f'), "sec") # 시간 출력(단위 : 초)
Checksort(In_l2, n)
print("증가식 : 2h + 1, 감소식 : h/2, N :", n)
print("Shell sort time : ", format(end_time2 - start_time2, 'f'), "sec") # 시간 출력(단위 : 초)
Checksort(In_l3, n)
print("증가식 : 5h + 1, 감소식 : h/5, N :", n)
print("Shell sort time : ", format(end_time3 - start_time3, 'f'), "sec") # 시간 출력(단위 : 초)
```
